### Total number of feedback,city wise

```python
df3=df[['city','rating_count']]
df3.drop_duplicates()
df4=df3.groupby(['city'])['rating_count'].sum()
df4=df4.to_frame()
df4=df4.sort_values('rating_count',ascending=False)
df4.head()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(df4.index,df4['rating_count'])
plt.xticks(rotation=90);
```

### Top cusines of india city wise

```python
df5=df[['cusine','rating']]
df5.drop_duplicates()
df6=df5.groupby(['cusine'])['rating'].sum()
df6=df6.to_frame()
df6=df6.sort_values('rating',ascending=False)
df6.head()
```
